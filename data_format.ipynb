```python
data = {
    "name": "palak",
    "age": 20
}

print(data)
```

```python
import json

## serialze
json_data = json.dumps(data)
print(json_data)
```

```python
# save
with open("data.json","w") as f:
  f.write(json_data)
  print("data stored file")
```

```python
with open("data.json","r") as f:
  data = json.load(f)
  print(data)
  print("data loaded")
```

```python
binary_data= bytes("json_data","utf8")
print(binary_data)
print(type(binary_data))
```

```python
data = {
    "name": "sargam",
    "age": 20,
    "city" : "shimla"
}

print(type(data))
```

```python
import json
json_data=json.dumps(data)
print(json_data)
print(type(json_data))
```

```python
## memory 
import sys

size_json=sys.getsizeof(json_data) # byte
print(size_json)
```

```python
# binary 
binary_data = bytes(json_data, "utf8")
print(binary_data)
print(type(binary_data))
```

```python
#memory size 
size_binary = sys.getsizeof(binary_data)
print(size_binary)
```
